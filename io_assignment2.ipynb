```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('housing.csv')
x = dataset
```

```python
print(x)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x1 = sc.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
x_new = pca.fit_transform(x1)
```

```python
print(x_new)
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x_new)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(x_new)
```

```python
plt.scatter(x_new[y_kmeans==0, 0], x_new[y_kmeans == 0, 1], s = 30, c = 'red', label = 'Cluster 1')
plt.scatter(x_new[y_kmeans==1, 0], x_new[y_kmeans == 1, 1], s = 30, c = 'blue', label = 'Cluster 2')
plt.scatter(x_new[y_kmeans==2, 0], x_new[y_kmeans == 2, 1], s = 30, c = 'green', label = 'Cluster 3')
plt.scatter(x_new[y_kmeans==3, 0], x_new[y_kmeans == 3, 1], s = 30, c = 'cyan', label = 'Cluster 4')
plt.scatter(x_new[y_kmeans==4, 0], x_new[y_kmeans == 4, 1], s = 30, c = 'pink', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 20, c = 'yellow', label = 'centroids')
plt.title('Clusters')
plt.xlabel('MedHouseVal')
plt.ylabel('MedInc')
plt.legend()
plt.show()
```
